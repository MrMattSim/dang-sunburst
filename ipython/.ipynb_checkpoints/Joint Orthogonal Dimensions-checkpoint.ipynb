```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate discrete joint distribution

N = 1500
x = np.random.normal(3,1,N)
y = stats.gamma(3).rvs(N)

H,xedges,yedges = np.histogram2d(y,x,[10,10])
X, Y = np.meshgrid(xedges,yedges)
```

```python
mpl.rc("figure", figsize=(6, 6))
ax = sns.kdeplot(x,y,shaded=True)
ax.set_xlim(0,10)
ax.set_ylim(0,6)
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
mpl.rc("figure", figsize=(6, 6))

pcolormesh(X,Y,H.T);
ax = gca();
ax.set_xlim([0,10])
ax.set_ylim([0,6])
ax.set_xlabel('X');
ax.set_ylabel('Y');
```

```python
import json

fd = []

nrows, ncols = shape(H)
for r in range(nrows):
    for c in range(ncols):
        #print "(%d,%d) = %0.2f"%(r,c,H[r][c])
        
        d = {}
        d['x'] = r+1
        d['y'] = c+1
        d['freq'] = H[r][c]/N
        fd.append(d)

with open('sunburst_orthogonal_xy.json','w') as f:
    json.dump(fd,f,sort_keys=True,indent=4,separators=(',', ': '))

print "Done printing flat JSON to file."
```

```python
tree = {}

# if we were more slick we'd do this recursively, 
# but we're not, so we won't.

tree['name'] = 'root'

children = []
for r in range(nrows):
    child = {}
    child['name'] = 'X%d'%(r)
    
    grandchildren = []
    for c in range(ncols):        
        grandchild = {}
        grandchild['name'] = 'X%d-Y%d'%(r,c)
        grandchild['freq'] = H[r][c]/N

        grandchildren.append(grandchild)
    
    child['children'] = grandchildren
    child['freq'] = sum(H[r][:])/N
    
    children.append(child)

tree['children'] = children
tree['freq'] = sum(H[:][:])/N


with open('sunburst_orthogonal_tree.json','w') as f:
    json.dump(tree,f,sort_keys=True,indent=4,separators=(',', ': '))

print "Done printing tree JSON to file."
```

```python
from numpy.random import weibull
from scipy.stats import cumfreq

Nsamples = 1e4

# generate a weibull distribution
k=5
x = weibull(k,(Nsamples,))
x2 = stats.gamma(2).rvs(Nsamples)
x3 = stats.gamma(2).rvs(Nsamples)

# now bin x1 into bins
Nbins = 8

counts, binlocs = np.histogramdd([x,x2,x3]);
#counts = 10x10x10
```

```python
tree = {}
tree['name'] = 'root'
tree['freq'] = sum(counts[:][:][:])/Nsamples

dims = shape(counts)
children1 = []
for d1 in range(dims[0]):
    child1 = {}
    child1['name'] = 'X%d'%(d1+1)
    child1['freq'] = sum(counts[d1][:][:])/Nsamples
    
    children2 = []
    for d2 in range(dims[1]):        
        child2 = {}
        child2['name'] = 'X%d-Y%d'%(d1+1,d2+1)
        child2['freq'] = sum(counts[d1][d2][:])/Nsamples
        
        children3 = []
        for d3 in range(dims[2]):
            
            child3 = {}
            child3['name'] = 'X%d-Y%d-Z%d'%(d1+1,d2+1,d3+1)
            child3['freq'] = sum(counts[d1][d2][d3])/Nsamples
            
            children3.append(child3)
            
        child2['children'] = children3            
        children2.append(child2)

    child1['children'] = children2
    children1.append(child1)

tree['children'] = children1

print tree

with open('sunburst_orthogonal2_tree.json','w') as f:
    json.dump(tree,f,sort_keys=True,indent=4,separators=(',', ': '))

print "Done printing tree JSON to file."
```
